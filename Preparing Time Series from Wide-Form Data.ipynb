# Import Libraries

```python
## Our standard imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import datetime as dt

#import tick customization tools
import matplotlib.ticker as mticks
import matplotlib.dates as mdates

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Data/City_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv')
df.head()
```

```python
df.info()
```

```python
# Filter the 4 largest cities into a new dataframe
df_4 = df.iloc[0:4]
df_4
```

```python
# Melt the data to long-form
melted_df = pd.melt(df_4, id_vars=['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', 'State', 'Metro', 'CountyName'], value_name='ZHVI', var_name='Date')
melted_df
```

```python
melted_df.info()
```

```python
# convert date to datetime dtype
melted_df['Datetime'] = pd.to_datetime(melted_df['Date'])
melted_df.info()
```

```python
melted_df.drop(columns='Date', inplace=True)
melted_df.head()
```

```python
# Make the datetime column the index
melted_df = melted_df.set_index('Datetime')
melted_df
```

```python
# Resample the dataframe as monthly frequency, grouped by City
df_ts = melted_df.groupby('RegionName').resample('M').sum()
df_ts
```

# Plot Home Values

```python
unstacked = df_ts.unstack(level=0)
unstacked
```

```python
def format_xdates(ax, fig=None, xtick_fontweight='bold',
               title= None):
    """Formats x-axis with minor ticks every 3 months, and major 
    ticks every year. Includes different grid options for visibility"""
    # create the locator to place ticks every 3 months.
    #loc_3months = mdates.MonthLocator(interval=3)
    #fmt_months = mdates.DateFormatter("%b")
    ## for major year ticks
    #loc_year = mdates.YearLocator()
    #fmt_year = mdates.DateFormatter("%Y")
    
    ## Change the font of the major ticks to stand out
    ax.set_xticks(ax.get_xticks())
    ax.set_xticklabels(ax.get_xticklabels(), 
                       fontweight=xtick_fontweight) 
    
    ## customize minor ticks
    #ax.xaxis.set_minor_locator(loc_3months)
    #ax.xaxis.set_minor_formatter(fmt_months)
    ## customize major ticks
    #ax.xaxis.set_major_locator(loc_year)
    #ax.xaxis.set_major_formatter(fmt_year)
    
    ## Making major/minor gridlines visually distinct
    ax.grid(which='minor',axis='x',ls=":")
    ax.grid(which='major',axis='x',color='k')
    if fig is None:
        fig = ax.get_figure()
    ## rotate the dates
    fig.autofmt_xdate(which='major',rotation=90,ha='center')
    return fig,ax
```

```python
def millions(x,pos):
    """function for use wth matplotlib FuncFormatter -  formats money in millions"""
    new_x = x/1_000_000
    return f"{new_x:,.0f}M"
def thousands(x,pos):
    """function for use wth matplotlib FuncFormatter -  formats money in millions"""
    new_x = x/1_000
    return f"{new_x:,.0f}K"
# Create the formatters
fmt_mill = mticks.FuncFormatter(millions)
fmt_thous = mticks.FuncFormatter(thousands)
```

```python
fig, ax = format_xdates(unstacked['ZHVI'].plot())

ax.yaxis.set_major_formatter(fmt_thous)

plt.title('Zillow House Value Index 2000-2022')
plt.ylabel('Value Index in $1,000s');
```
